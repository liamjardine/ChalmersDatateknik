Assumptions and Motivate decisions:

We choose to study the relationship between income inequality and trust towards others within society. After discussing the topic, we both landed in the assumption that intuitively it would seem to us that a society with a higher level of income inequality would have generally lower interpersonal trust.

We feel like there is a discourse in most industrialized nations about socioeconomic class as a way of categorizing people within nations. And a nation with a wider gap between these different groups could lead to common mistrust towards other people in general or at least towards other groups. Suspicion of corruption, nepotism or just a general mistrust towards other more "well of" groups but also the ither way around, downwards the socioeconomic ladder.

Questions we hope the answer through the data are:

   * Do assumptions hold? Do countries with higher income inequality have lower interpersonal trust?
   * Are there any outliers?
   * If so, how are these countries different from the global mean?

Some demarcation has been made to clean the data and narrow the field of search. We have excluded any data with non numerical value. This could result in some sample bias as some countries are better represented than others. Also the data used is from the last 30 years as some countries might only have more recent data or has in some way differed heavily from older records, much like the first assignment. We have excluded some columns from the dataframe which are of no interest in this report. We would again use a standard deviation of 0.3 or more to evaluate outliers for income inequality and interpersonal trust.

```python
#Load data_set paths into Dataframe
df_trust_income = pd.read_csv('C:\DEV\DataScienceAndAI\Jupyter\Assignment 1\datasets\interpersonal-trust-vs-income-inequality.csv')

# remove rendundant columns
df_trust_income = df_trust_income.iloc[:, [0, 2, 3, 4]]

# remove rendundant columns
df_trust_income = df_trust_income.dropna()

# remove any data older than 30 years
df_trust_income = df_trust_income[df_trust_income.Year >= 1992]

# reset index
df_trust_income.reset_index(drop=True, inplace=True)
```

```python
# Get the unique countries
unique_countries = list(df_trust_income['Entity'].drop_duplicates())

# Create new dataframe to store data to be plotted
new_df_trust_income = pd.DataFrame(unique_countries, columns = ['Country'])

# create required columns
new_df_trust_income['avg_trust'] = None
new_df_trust_income['avg_income'] = None

# loop to update new dataframe with the mean values of GDP, life expectancy
for i in range(len(list(unique_countries))):
    if new_df_trust_income.loc[i]['Country'] == unique_countries[i]:
        stats = df_trust_income[df_trust_income.Entity == unique_countries[i]].describe()
        new_df_trust_income.loc[i]['avg_trust'] = stats.loc['mean']['Trust in others (World Values Survey (2014))']
        new_df_trust_income.loc[i]['avg_income'] = stats.loc['mean']['Gini index']
```

```python
income = new_df_trust_income['avg_income']

trust = new_df_trust_income['avg_trust']

fig, ax = plt.subplots(figsize=(100, 50))
ax.scatter(x=income, y=trust, s=250)
ax.set_xlabel('Income inequality', size=100)
ax.set_ylabel('Trust in others', size=100)
ax.set_title('Trust in other vs. Income inequality', size=200)
ax.tick_params(axis='x', labelsize=100)
ax.tick_params(axis='y', labelsize=100)

for i, txt in enumerate(list(new_df_trust_income['Country'])):
        ax.annotate(txt, (new_df_trust_income.loc[i]['avg_income'], new_df_trust_income.loc[i]['avg_trust']), size=50)
```

One could imagine some regression line leaning from top left to bottom right. This would hold with the assumptions made about countries with lower income inequality would have a higher interpersonal trust and vice versa. There seems to be a few outliers although nothing too extreme.

Something that caught our eye is that the outliers in the bottom left are mostly east european countries. Here income inequality is low and so is also trust among others. Could there be a reason for this, or this an effect of a sample bias? Not all countries are represented equally in the data. Let's take a further look at Slovenia and The United States as examples.


```python
overall_stats = df_trust_income.describe()

overall_mean_trust = overall_stats.loc['mean']['Trust in others (World Values Survey (2014))']
overall_std_trust = overall_stats.loc['std']['Trust in others (World Values Survey (2014))']

overall_mean_income = overall_stats.loc['mean']['Gini index']
overall_std_income = overall_stats.loc['std']['Gini index']

high_trust_df = new_df_trust_income[new_df_trust_income['avg_trust'] > overall_mean_trust + overall_std_trust * 0.3]
low_trust_df = new_df_trust_income[new_df_trust_income['avg_trust'] < overall_mean_trust + overall_std_trust * 0.3]

high_income_high_trust_df = low_trust_df[low_trust_df['avg_income'] > overall_mean_income + overall_std_income * 0.3]
high_income_high_trust_df = low_income_high_trust_df.reset_index()

low_income_low_trust_df = low_trust_df[low_trust_df['avg_income'] < overall_mean_income + overall_std_income * 0.3]
low_income_low_trust_df = low_income_high_trust_df.reset_index()

low_income_low_trust_df
```

```python
fig, ([ax0, ax1], [ax2, ax3], [ax4, ax5]) = plt.subplots(nrows=3, ncols=2, figsize=(15, 15))

# Plot violin plot on axes 1

ax0.violinplot(df_trust_income['Gini index'], showmedians=True)
ax0.set_title('Income inequality globally')

# Plot violin plot on axes 2

ax1.violinplot(df_trust_income['Trust in others (World Values Survey (2014))'], showmedians=True)
ax1.set_title('Trust towards others globally')

# Plot violin plot on axes 3

df_trust_income_slovenia = df_trust_income[df_trust_income.Entity == 'Slovenia']

ax2.violinplot(df_trust_income_slovenia['Gini index'], showmedians=True)
ax2.set_title('Income inequality Slovenia')

# Plot violin plot on axes 4

ax3.violinplot(df_trust_income_slovenia['Trust in others (World Values Survey (2014))'], showmedians=True)
ax3.set_title('Trust towards others Slovenia')

# Plot violin plot on axes 3

df_trust_income_unitedstates = df_trust_income[df_trust_income.Entity == 'China']

ax4.violinplot(df_trust_income_unitedstates['Gini index'], showmedians=True)
ax4.set_title('Income inequality China')

# Plot violin plot on axes 4

ax5.violinplot(df_trust_income_unitedstates['Trust in others (World Values Survey (2014))'], showmedians=True)
ax5.set_title('Trust towards others China')

plt.show()
```

By using the violinplot we can give a clear vizualisation of the density distrubution of the data and mean. Here we see the global mean next to data from Slovenia and China.
