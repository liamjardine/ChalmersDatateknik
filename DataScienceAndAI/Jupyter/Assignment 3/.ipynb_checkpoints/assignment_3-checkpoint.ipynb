2. Use  the  K-means  clustering  method to  cluster  the  phi  and  psi  angle  combinations  in  
the data file.

a. Experiment with different values of K. Suggest an appropriate value of K for this 
task and motivate this choice.

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Perform K-Means clustering     
# Specify the number of clusters and fit the data X 

kmeans = KMeans(n_clusters=4, random_state=0).fit(df_phi_psi)
print(kmeans.cluster_centers_)

fig, ax = plt.subplots(figsize=(70, 50))

ax.set_title('Distribution of Î¦ (phi) and Î¨ (psi)', size=100)
ax.set_xlabel('Î¦ (phi)', size=75)
ax.set_ylabel('Î¨ (psi)', size=75)
ax.tick_params(axis='x', labelsize=75)
ax.tick_params(axis='y', labelsize=75)

array = df_phi_psi.to_numpy()
x = array[:, 0]
y = array[:, 1]
ax.scatter(x, y)
ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=1000)
```

```python
distortions = []
K = range(1,11)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(df_phi_psi)
    distortions.append(kmeanModel.inertia_)
     
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

b. Validate the clusters that are found with the chosen value of K.

We can validate the clustering using by checking the stability on the subsets, check for Co-occurence, the stability over repetitions and the silhouette coefficient. We will go through these step-by-step.

The stability on subsets is controlled by removing a proportion of some random points to see if clusters remain fundamentally the same.
